## Part 2

Write a block of commands that does the following.

## Part 3

1. Create an identity matrix called `identity_matrix` that has the same size as `img_average_channel`.

```python
# Get the shape of the grayscale image
img_shape = imgAverageChannel.shape

# Create an identity matrix with the same shape as the grayscale image
identityMatrix = np.eye(img_shape[0], img_shape[1])

# Generate a random row index within the range of the image height
random_row_index = np.random.randint(0, img_shape[0])

# Extract a random row from the identity matrix
random_row = identityMatrix[random_row_index, :]

# Print the identity matrix
print(identityMatrix)

# Print the randomly selected row from the identity matrix along with its index
print("\nRandom Row from Identity Matrix (Row Index {}):".format(random_row_index))
print(random_row)
```

## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`. 

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# Get the number of rows in the grayscale image
num_rows = imgAverageChannel.shape[0]

# Create an array containing all row indices from 0 to (num_rows-1)
rows = np.arange(num_rows)

# Shuffle the row indices randomly
np.random.shuffle(rows)

# Print the shuffled row indices
print("Shuffled Rows:", rows)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
# Select rows from the identity matrix according to the shuffled row indices
T = identityMatrix[rows]

# Print the shape of the resulting matrix T
print("Shape of T:", T.shape)

# Check if the shape of T is different from the shape of the grayscale image
if T.shape != imgAverageChannel.shape:
    # Reshape T to match the shape of the grayscale image if needed
    T = T.reshape(imgAverageChannel.shape)

# Print the shape of T after potential reshaping
print("Reshaped T Shape:", T.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
# Perform a matrix multiplication between the shuffled identity matrix T and the grayscale image
transformed_image = np.dot(T, imgAverageChannel)

# Print the shape of the transformed image
print("Shape of transformed_image:", transformed_image.shape)

# Display the transformed image using a grayscale colormap
plt.imshow(transformed_image, cmap='gray')

# Set a title for the displayed image
plt.title("Transformed Image")

# Show the image plot
plt.show()
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
# Check if matrix T is invertible by calculating its determinant
det_T = np.linalg.det(T)

if det_T != 0:
    # If T is invertible, compute its inverse
    Tinv = np.linalg.inv(T)

    # Perform the inverse operation to recover the original image
    original_image = np.dot(Tinv, transformed_image)

    # Plot the original_image
    plt.imshow(original_image, cmap='gray')
    plt.title("Original Image (Recovered)")
    plt.show()
else:
    print("Matrix T is not invertible. Determinant is zero.")

    
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Get the number of rows in the image
num_rows = image.shape[0]

# Create an array containing all row indices from 0 to (num_rows-1)
rows = np.arange(num_rows)

# Shuffle the row indices randomly
np.random.shuffle(rows)

# Get the number of columns in the image
num_cols = image.shape[1]

# Create an array containing all column indices from 0 to (num_cols-1)
cols = np.arange(num_cols)

# Shuffle the column indices randomly
np.random.shuffle(cols)

# Shuffle both rows and columns of the image to create S_image
S_image = image[rows, :, :]
S_image = S_image[:, cols, :]

# Display the shuffled image
plt.imshow(S_image)
plt.title("Shuffled Image (Both Rows and Columns)")
plt.show()

# Use argsort to obtain the indices that would sort the shuffled rows and columns back to their original order
I_rows = np.argsort(rows)
I_cols = np.argsort(cols)

# Recover the original image by sorting rows and columns using the obtained indices
R_image = S_image[I_rows, :, :]
R_image = R_image[:, I_cols, :]

# Display the recovered original image
plt.imshow(R_image)
plt.title("Recovered Original Image")
plt.show()
```
